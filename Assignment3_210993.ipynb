```python
import yfinance as yf
import matplotlib.pyplot as plt
import mplfinance as mpf



stock_symbols = ["AMZN", "AAPL", "GOOGL", "MSFT", "TSLA", "GOOG", "NFLX", "NVDA", "AMGN", "JPM"]





for stock_symbol in stock_symbols:
   

    data = yf.download(stock_symbol, period="10y", interval="1d")

    
    mpf.plot(data, type='candle', style='charles')

    
    pattern_observed = False

    # Ascending triangle pattern
    ascending_triangle_start = None

    for i in range(2, len(data)):
        if data['High'][i] <= data['High'][i-1] and data['Low'][i] > data['Low'][i-1]:
            if ascending_triangle_start is None:
                ascending_triangle_start = data.index[i-1]
                pattern_observed = True
                break

    
    if pattern_observed:
        print(f"{stock_symbol} - Ascending Triangle Pattern Start Date: {ascending_triangle_start}")

        
        
        mpf.plot(data.loc[ascending_triangle_start:], type='candle', style='charles', title='Ascending Triangle Pattern')
    else:
        print(f"{stock_symbol} - Ascending Triangle Pattern Not Observed")

    
    pattern_observed = False

    # Bear flag pattern
    bear_flag_start = None

    for i in range(2, len(data)):
        if data['High'][i] < data['High'][i-1] and data['Low'][i] > data['Low'][i-1] and data['Close'][i] < data['Open'][i-1]:
            if bear_flag_start is None:
                bear_flag_start = data.index[i-2]
                pattern_observed = True
                break

   
    if pattern_observed:
        print(f"{stock_symbol} - Bear Flag Pattern Start Date: {bear_flag_start}")
        
        mpf.plot(data.loc[bear_flag_start:], type='candle', style='charles', title='Bear Flag Pattern')
    else:
        print(f"{stock_symbol} - Bear Flag Pattern Not Observed")

   
    pattern_observed = False

    # Cup and handle pattern
    cup_start = None

    for i in range(2, len(data)):
        if data['High'][i] <= data['High'][i-1] and data['Low'][i] >= data['Low'][i-1] and data['High'][i] < data['High'][i-2] and data['Low'][i] > data['Low'][i-2]:
            if cup_start is None:
                cup_start = data.index[i-2]
                pattern_observed = True
                break


                
    if pattern_observed:
        print(f"{stock_symbol} - Cup and Handle Pattern Start Date: {cup_start}")
        
                
        mpf.plot(data.loc[cup_start:], type='candle', style='charles', title='Cup and Handle Pattern')
    else:
        print(f"{stock_symbol} - Cup and Handle Pattern Not Observed")

    
    pattern_observed = False

    # Rounding top pattern
    rounding_top_start = None

    for i in range(3, len(data)):
        if (
            data['High'][i] <= data['High'][i-1]
            and data['High'][i-1] >= data['High'][i-2]
            and data['High'][i-2] <= data['High'][i-3]
            and data['Low'][i] <= data['Low'][i-1]
            and data['Low'][i-1] <= data['Low'][i-2]
            and data['Low'][i-2] <= data['Low'][i-3]
        ):
            if rounding_top_start is None:
                rounding_top_start = data.index[i-3]
                pattern_observed = True
                break

    
    if pattern_observed:
        print(f"{stock_symbol} - Rounding Top Pattern Start Date: {rounding_top_start}")

        mpf.plot(data.loc[rounding_top_start:], type='candle', style='charles', title='Rounding Top Pattern')
    else:
        print(f"{stock_symbol} - Rounding Top Pattern Not Observed")

```
